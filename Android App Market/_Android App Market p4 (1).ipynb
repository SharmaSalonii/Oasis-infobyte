#### Finding and Removing Outliers 

```python
%matplotlib inline
```

```python
sns.boxplot(df1["Rating"])
plt.show()
```

```python
Q1 = df1['Rating'].quantile(0.25)
Q3 = df1['Rating'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_outliers = df1[(df1['Rating'] >= lower_bound) & (df1['Rating'] <= upper_bound)]

print("Original dataset shape:", df1.shape)
print("Dataset shape after removing outliers:", df_outliers.shape)
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(df_outliers['Rating'])
plt.title('Rating Distribution without Outliers')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(y='Category', data=df1, order=df1['Category'].value_counts().index)
plt.title('Number of Apps in Each Category')
plt.show()
```

```python
sns.histplot(df1['Rating'], bins=10, kde=True)
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
labels = df1['Type'].value_counts().index
sizes = df1['Type'].value_counts().values
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Free vs Paid Apps Distribution')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Category', y='Rating', data=df1)
plt.xticks(rotation=90)
plt.title('Ratings Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.show()
```
