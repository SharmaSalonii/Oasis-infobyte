#### Finding and Removing Outliers 

```python
sns.boxplot(df3["Rating"])
plt.show()
```

```python
Q1 = df3['Rating'].quantile(0.25)
Q3 = df3['Rating'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_outliers = df3[(df3['Rating'] >= lower_bound) & (df3['Rating'] <= upper_bound)]

print("Original dataset shape:", df3.shape)
print("Dataset shape after removing outliers:", df_outliers.shape)
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(df_outliers['Rating'])
plt.title('Rating Distribution without Outliers')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(y='Category', data=df3, order=df3['Category'].value_counts().index)
plt.title('Number of Apps in Each Category')
plt.show()
```

```python
sns.histplot(df3['Rating'], bins=10, kde=True)
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
labels = df3['Type'].value_counts().index
sizes = df3['Type'].value_counts().values
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Free vs Paid Apps Distribution')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Category', y='Rating', data=df3)
plt.xticks(rotation=90)
plt.title('Ratings Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.show()
```

##### Do paid apps have higher ratings than free apps?


```python
sns.barplot(x='Type', y='Rating', data=df3)
plt.title('Ratings of Free vs Paid Apps')
plt.xlabel('App Type (Free = 1, Paid = 0)')
plt.ylabel('Rating')
plt.show()
```

##### number of reviews affect app ratings

```python
sns.scatterplot(x='Reviews', y='Rating', data=df3)
plt.title('Reviews vs Rating')
plt.xlabel('Number of Reviews')
plt.ylabel('Rating')
plt.show()
```

```python
paid_apps = df3[df3['Type'] == 0]
plt.figure(figsize=(12, 8))
sns.countplot(y='Category', data=paid_apps, order=paid_apps['Category'].value_counts().index)
plt.title('Number of Paid Apps per Category')
plt.ylabel('Category')
plt.xlabel('Number of Paid Apps')
plt.show()
```
