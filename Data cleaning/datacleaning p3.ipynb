### Handling missing values

```python
df["last_review"] = pd.to_datetime(df["last_review"], dayfirst=True, errors='coerce')
```

```python
df['last_review'] = df['last_review'].ffill()
df.ffill(inplace=True)
```

```python
df.dropna(subset=["host_name"],inplace=True)
```

```python
df['reviews_per_month'].fillna(0,inplace=True)
```

```python
df.dropna(subset=["name"],inplace=True)
```

### Finding and removing outliers

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_cleaned = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
```

### Standardscaler only numeric columns 

##### StandardScaler(): This calculates the mean and standard deviation of each feature and scales the data accordingly.

```python
from sklearn.preprocessing import StandardScaler

numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
print(df)
```
