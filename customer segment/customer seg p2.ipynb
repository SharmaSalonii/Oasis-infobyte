#### Outliers detect and remove

```python
plt.figure(figsize=(6, 4))  
sns.boxplot(data=df, y='MntTotal')
plt.title('Box Plot for MntTotal')
plt.ylabel('MntTotal')
plt.show()
```

```python
Q1 = df['MntTotal'].quantile(0.25)
Q3 = df['MntTotal'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['MntTotal'] < lower_bound) | (df['MntTotal'] > upper_bound)]
outliers.head()
```

```python
data = df[(df['MntTotal'] > lower_bound) & (df['MntTotal'] < upper_bound)]
data.describe()
```

```python
data = data[(data['MntTotal'] > lower_bound) & (data['MntTotal'] < upper_bound)]
data.describe()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3,n_init=10)
df['Cluster'] = kmeans.fit_predict(df[['Age', 'Income']])

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Age', y='Income', hue='Cluster', palette='Set2')
plt.title('Customer Segments by Age and Income')
plt.show()
```

```python
def get_marital_status(row):
    if row['marital_Divorced'] == 1:
        return 'Divorced'
    elif row['marital_Married'] == 1:
        return 'Married'
    elif row['marital_Single'] == 1:
        return 'Single'
    elif row['marital_Together'] == 1:
        return 'Together'
    elif row['marital_Widow'] == 1:
        return 'Widow'
    else:
        return 'Unknown'
data['Marital'] = data.apply(get_marital_status, axis=1)
```

```python
sns.barplot(data=data, x='Education', y='MntTotal',hue="Marital", palette='muted') 
plt.title('Spending by Marital Status')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Marital', y='MntTotal', data=data, palette='muted')
plt.title('MntTotal by marital status')
plt.xlabel('Marital status')
plt.ylabel('MntTotal')
```

```python
sns.boxplot(data=data, x='Marital', y='MntTotal', palette='muted') 
plt.title('Spending by Marital Status')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Education', y='MntWines', data=data, hue="Marital")

plt.title('Wine Spending by Education Level and Marital Status')
plt.xlabel('Education')
plt.ylabel('Amount Spent on Wine')
plt.show()
```
